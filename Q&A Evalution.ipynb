Ideal Answer: """

```python
prompt="""
You are an Evaluator. Your task is to assess student answers to exam questions based on a specific rubric.

Rubric:

Accuracy (A):
      - Assess how accurately the student's answer addresses the question.
      - Compare it to the ideal answer provided.
      - Assign a score from 0 (lowest accuracy) to 1 (highest accuracy) to the variable 'A'.

Keywords (K):
      - Identify primary and secondary keywords expected in the answer.
      - Compare the presence and usage of these keywords in both the student's answer and the ideal answer.
      - Assign a score from 0 to 1 to the variable 'K' based on keyword coverage and relevance.

Grammar and Punctuation (G):
      - Evaluate the student's answer for grammatical accuracy and proper punctuation.
      - Assign a score from 0 (most errors) to 1 (no errors) to the variable 'G'.

Scoring:
      - Calculate the average of scores A, K, and G.
      - Multiply the average by 10 to obtain a final score out of 10.

In this case question us based in English Poem. Therefore we are giving out poem to you as well.

Poem:
The Sun's Reflection
The sun's reflection, caught in puddles deep,
A shimmered jewel where secrets softly sleep.
It holds the sky, like canvas stretched on blue,
But ripples distort, and dreamscape paints anew.

Is this the world, reflected, fractured, strange?
Or just a fleeting glimpse, a passing change?
Does truth lie nestled in its depths unseen,
Or is it simply light, a play on mirrored sheen?

Question:///
In 'The Sun's Reflection,' the poet plays with the image of a puddle reflecting the sky. How does this image contribute to the poem's central theme? Consider elements like form, language, and symbolism in your answer.
///

Ideal Answer: ///
The puddle serves as a powerful symbol for a world filled with uncertainty and fragmented perspectives. The distorted reflection of the sky challenges us to question whether we perceive reality accurately or merely see our own projections. The use of short, choppy lines mirrors the ripples in the water, further emphasizing the fragmented nature of our experience. The poem's ambiguity, with its unanswered questions, invites us to contemplate the relationship between truth, illusion, and our own internal landscapes.
///

Student Answer: {student_answer}

Now compare the students answer with ideal answer and assign suitable score for it.
"""
```

```python
student_answer="""
The puddle simply reflects the sky, showing us the beauty of nature. The poem is a call to appreciate the ordinary things around us.
"""
```

```python
prompt_template=PromptTemplate(template=prompt, input_variables=['student_answer'])
chain=LLMChain(llm=llmchat, prompt=prompt_template, verbose=True)
overall_chain = SimpleSequentialChain(chains=[chain], verbose=True)
answer=overall_chain.run(student_answer)
print(answer)
```
